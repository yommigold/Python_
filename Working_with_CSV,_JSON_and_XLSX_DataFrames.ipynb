```python
# Working with Jay.Json File

#1) Download Jay.json in C7W1 Activity folder
#2) Import the Json in dataframe
#3) Add a column to the dataframe
#4) Perform describe and info on dataframe
#5) Export and save this data as Jay1.json
```

```python
# Import the Json in dataframe

import pandas as pd
import json

with open('Jay.json') as f:
    data = json.load(f)
df = pd.DataFrame(data)
df
```

```python
# Add a column to the dataframe

df["Sugar_Level"] = 200 # with all values initially set to 200
df
```

```python
#Perform describe on dataframe

df.describe()
```

```python
#  Perform info on dataframe

df.info()
```

```python
# Export and save this data as Jay1.json

df.to_json('Jay1.json')
df
```
