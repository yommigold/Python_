```python
# The functions below will download the dataset into your browser and its optional
!pip install pyodide
import pyodide

from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
```

```python
import pycurl
```

```python
file_path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
```

```python
def download(file_path, url):
    c = pycurl.Curl()
    c.setopt(c.URL, url)
    c.setopt(c.WRITEDATA, open(file_path, "wb"))
    c.perform()
    c.close()
```

```python
df = pd.read_csv(file_path)
```

```python
# show the first 5 rows using dataframe.head() method
print("The first 5 rows of the dataframe")
df.head(5)
```

```python
# Question 1: Check the bottom 10 rows of the DataFrame
print("The last 10 rows of the dataframe\n")
df.tail(10)
```

```python
# Add Headers

# Thus, you have to add headers manually
# First, create a list "headers" that include all column names in order
# Then, use dataframe.columns = headers to replace the headers with the list you created
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# Replace headers and recheck our data frame:
df.columns = headers
df.columns
```

```python
# You can also see the first 10 entries of the updated data frame and note that the headers are updated.

df.head(10)
```

```python
# Now, we need to replace the "?" symbol with NaN so the dropna() can remove the missing values:
df1=df.replace('?',np.NaN)
```

```python
# You can drop missing values along the column "price" as follows:
df=df1.dropna(subset=["price"], axis=0)
df.head(20)

# Here, axis=0 means that the contents along the entire row will be dropped wherever the entity 'price' is found to be NaN
```

```python
# Question 2: Find the name of the columns of the dataframe

print(df.columns)
```

```python
# Save Dataset
# Correspondingly, Pandas enables you to save the data set to CSV. By using the dataframe.to_csv() method
# You can add the file path and name along with quotation marks in the brackets

df.to_csv("automobile.csv", index=False)

# index = False means the row names will not be written
```
