# Example 1:
The Cal Poly course catalog (http://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory) contains a list of courses offered by the Statistics department. Scrape this website to obtain a DataFrame, where the unit of observation is a course. Use this DataFrame to answer the following questions:



How many 300-level courses does the Statistics department offer?

```python
#Step 1:Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Step 2:Adding URL
response = requests.get("https://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory")
response
```

```python
#Step 3:Creating Soup object
soup = BeautifulSoup(response.text,"html.parser")
soup
```

```python
#Step 4: Retrieve the Data
#soup.div.class_="courseblocktitle"
soup.find(class_="courseblocktitle").text
```

```python
List1=[]

for i in soup.find_all(class_="courseblocktitle"):
   text2 = i.text[0:8]
   List1.append(text2)

List1
```

```python
import pandas as pd
df = pd.DataFrame(List1, columns=['Coursecode'])
df
```

```python
# see this for guide #https://pandas.pydata.org/docs/user_guide/text.html

df2 = df[df['Coursecode'].str.contains('3')]
df2.head()
```

```python
# How many 300-level courses does the Statistics department offer?
df2 = df[df['Coursecode'].str.get(5)=='3']
df2.head()
```

Example 2:
Collect all the tags of the professor from this url: https://www.ratemyprofessors.com/professor/941931

```python
#Step 1:Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Step 2:Adding URL
# Send a GET request to the URL
url = "https://www.ratemyprofessors.com/professor/941931"
response = requests.get(url)
response
```

```python
#Step 3:Creating Soup object
# Parse the html content
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
#Step 4: Retrieve the Data
# Find all the tags associated with the professor
tags = soup.find_all(class_='Tag-bs9vf4-0')
tags
```

```python
# Extract the text from the tags
tag_list = [tag.text.strip() for tag in tags]
```

```python
# Print the list of tags
print("Tags associated with the professor:")
for tag in tag_list:
    print(tag)
```

```python
import requests
from bs4 import BeautifulSoup

def get_professor_tags(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        tags = soup.find_all(class_='Tag-bs9vf4-0')
        return [tag.text for tag in tags]
    except Exception as e:
        print(f"Error fetching tags: {e}")
        return []

professor_url = "https://www.ratemyprofessors.com/professor/941931"
tags = get_professor_tags(professor_url)

if tags:
    print(f"Tags for the professor: {', '.join(tags)}")
else:
    print("No tags found for this professor.")
```
